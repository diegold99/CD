```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
#Se usar o arquivo descompactado
dados=pd.read_csv('DOM2015.csv',sep=',')

#Se usar o arquivo compactado como está no dropbox púbico
#pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
#RENDA DA CLASSE C (+- 4000 A 9000)

dados=dados.loc[(dados.V4621>=4000) & (dados.V4621<=9000),['V0207','V0220','V0231','V0232','V02324','V02424','V02426']] #define colunas


dados


```

```python


dados.V0207 = dados.V0207.astype("category")
dados.V0220 = dados.V0220.astype("category")
dados.V0231 = dados.V0231.astype("category")
dados.V0232 = dados.V0232.astype("category")
dados.V02324 = dados.V02324.astype("category")
dados.V02424 = dados.V02424.astype("category")
dados.V02426 = dados.V02426.astype("category")


dados.V0207.cat.categories = ("Próprio-Pago","Próprio-NPago","Alugado","Cedido-Empregador","Cedido-Outros","Outra Condição")
dados.V0220.cat.categories = ("Sim","Nao")
dados.V0231.cat.categories = ("Sim","Nao")
dados.V0232.cat.categories = ("Sim","Não")
dados.V02324.cat.categories = ("Sim","Não")
dados.V02424.cat.categories = ("Sim","Não")
dados.V02426.cat.categories = ("Sim","Não")

dados
```

```python
dados = dados.rename(columns={"V0207":"Ocupação","V0220": "Celular","V0231":"Microcomputador","V0232":"Internet no Comp","V02324":"Internet no Celular","V02424":"3G ou 4G","V02426":"Banda Larga"})
dados = dados[pd.notnull(dados["Ocupação"])]

dados
```
